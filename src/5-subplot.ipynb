### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart (Top-Left)
axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('x')
axs[0, 0].grid(True)

# Bar chart (Top-Right)
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot (Bottom-Left)
axs[1, 0].scatter(x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid(True)

# Pie chart (Bottom-Right)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio for perfect circle

# Layout adjustment
plt.tight_layout()
plt.show()
```
