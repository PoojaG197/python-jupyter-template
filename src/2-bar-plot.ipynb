### Exercise 2: Bar Plot with Colors

```python

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']  # Custom colors for each bar

# Plotting
plt.figure(figsize=(7, 5))
plt.bar(categories, values, color=colors)

# Customizations
plt.title('Bar Plot of Categories A to D')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```
